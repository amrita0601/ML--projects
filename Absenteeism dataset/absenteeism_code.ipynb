We will use logistic regression as our classifying model. we will have 2 classes:
a)excessively absent
b)moderately absent
we will have the mean value of target variable as a cutoff. Values above median will fall into class a and less than median into class b.

```python
#finding median
med= data['Absenteeism Time in Hours'].median()
med
```

```python
#creating the 2 class
target=np.where(data['Absenteeism Time in Hours']>med,1,0)
target
```

```python
#adding it to dataset
data['excessively absent']=target
data.head()
```

```python
#dropping the absenteeism column
#dropping the next 3 columns as a later change
data=data.drop(['Absenteeism Time in Hours','Week_day','Distance to Work','Daily Work Load Average'],axis=1)
data.head()
```

```python
#defining inputs
unscaled_inputs=data.iloc[:,:-1]
unscaled_inputs.head()
```

```python
#importing lib for scalling
#from sklearn.preprocessing import StandardScaler
```

```python
#correcting mistake of stnd all inputs
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator,TransformerMixin):
    
    def __init__(self,columns,copy=True,with_mean=True,with_std=True):
        self.scaler=StandardScaler(copy,with_mean,with_std)
        self.columns=columns
        self.mean_=None
        self.var_=None
    
    def fit(self,X,y=None):
        self.scaler.fit(X[self.columns],y)
        self.mean_=np.mean(X[self.columns])
        self.var_=np.var(X[self.columns])
        return self
        
    def transform(self,X,y=None,copy=None):
        init_col_order= X.columns
        X_scaled=pd.DataFrame(self.scaler.transform(X[self.columns]),columns=self.columns)
        X_not_scaled=X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled,X_scaled],axis=1)[init_col_order]
```

```python
#syndardising only non-dummy columns
#columns_to_scale=['List_month','Week_day', 'Transportation Expense', 'Distance to Work', 'Age',
 #      'Daily Work Load Average', 'Body Mass Index', 'Children', 'Pets']
columns_to_omit=['reason_1','reason_2','reason_3','reason_4','Education']
columns_to_scale=[x for x in unscaled_inputs.columns.values if x not in columns_to_omit]
absenteeism_scaler=CustomScaler(columns_to_scale)
absenteeism_scaler.fit(unscaled_inputs)
inputs=absenteeism_scaler.transform(unscaled_inputs)
```

```python
#importing lib
from sklearn.model_selection import train_test_split
```

```python
#splitting the dataset
x_train,x_test,y_train,y_test= train_test_split(inputs,target,train_size=0.8,random_state=20)
```

```python
#checking shapes
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
#import lib
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
#train the model
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
#calculating accuracy of training data
lr.score(x_train,y_train)
```

```python
#finding intercepts and coeff
print(lr.intercept_)
lr.coef_
```

```python
#creating summary table
feature_names=['reason_1', 'reason_2', 'reason_3', 'reason_4', 'List_month',
        'Transportation Expense',  'Age',
       'Body Mass Index', 'Education', 'Children',
       'Pets']
summary_table=pd.DataFrame(data=feature_names,columns=['Feature_names'])
summary_table['Coefficients']=np.transpose(lr.coef_)
summary_table
```

```python
#adding the intercept
summary_table.index=summary_table.index+1 #shifting index by 1
summary_table.loc[0]=['Intercept',lr.intercept_[0]]
summary_table=summary_table.sort_index()
summary_table
```

Here, in logistic regression eqn of the best fitted line is:
log(odds)=b0+x1b1+...
=>odds= exponential of the above rhs
The above table shows  the coeff of log(odds)
lets find the coeff of odds

```python
#adding odds coeff to above table
summary_table['Odds_coeff']= np.exp(summary_table['Coefficients'])
summary_table
```

A feature is not imp if :
a) it's coeff is around 0
b) its odds ratio is around 1
A coeff of 0 implies that no matter the feature value, we will multiply it by 0
For a unit change in the stand feature, the odds inc by a multiple equal to the odds ratio(1= nochange)
